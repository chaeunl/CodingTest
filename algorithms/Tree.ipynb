# Tree

* 정의: # of nodes $\geq 1$
* 표현: root + $T_1,...,T_n$ (sub-trees)
* 용어:
 * root, siblings, ancestors (현재 노드에서 root 까지 모두), sub-tree
 * degree: node 에 연결 된 siblings 의 수, degree of tree: max(degrees)
 * height: root node 에서 node 에 이르는 edge 의 수, height of tree: max(heights)
 * level = height
 


## 1) Binary Tree

* 정의: 모든 nodes 가 2개의 sub-trees 를 갖는 tree
* level i 에서 최대 노드 수 = $2^i$
* $h+1 \leq$ # of nodes of BT $\leq 2^{h+1}-1$ (단, hegiht of tree=h)
* 종류:
 * Full Binary Tree: # of nodes of BT = $2^{h+1}-1$
 * Complete Binary Tree: 빠짐없이 왼쪽부터 차곡차곡 쌓은 것. $$
 * Skewed Binary Tree: 왼쪽 혹은 오른쪽에만 node 가 존재. # of nodes of BT = $h+1$

```python
class Queue:
    def __init__(self):
        self.head, self.tail = -1, -1
        self.size = 0
        self.queue = []
        
    def enqueue(self, data):
        self.queue.append(data)
        self.size += 1
        self.tail += 1
        
    def dequeue(self):
        if self.size < 1:
            print("Fucking queue!!")
        else:
            data = self.queue.pop(0)
            self.size -= 1
            self.tail -= 1
        return data
    
    def is_empty(self):
        return self.size < 1
```

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        
```

## 2) Binary Search Tree

<div></div>

* 모든 nodes 가 서로 다른 key 를 가짐: Left sub-tree < root < right sub-tree
* Inorder traversal $\rightarrow$ 오름차순 정렬
* time complexity: O(logN) (average) O(N) (worst case)

```python
class BinarySearchTree(BinaryTree):
    def __init__(self):
        super().__init__()   
```

### 2-1) search

1. root 에서 시작
2. x = root or x < root $\rightarrow$ left sub-tree search or x > root $\rightarrow$ right sub-tree search 

<div></div>


```python
def _search(self, x, curr_node):
    if (curr_node == None):
        return curr_node
    else:
        if x == curr_node.key:
            return curr_node
        elif x > curr_node.key:
            return self._search(x, curr_node.left)
        else:
            return self._search(x, curr_node.right)
    
def search(self, x):
    return self._search(x, self.root)
    
BinarySearchTree._search = _search
BinarySearchTree.search = search
```

### 2-2) insertion

1. 같은 x가 있는지 먼저 search
2. 탐색 실패한 위치에 x 삽입/탐색 성공한 위치의 x 삭제

<div></div>


```python
def _insert(self, x, curr_node):
    if (curr_node.key == x):
        return False
    else:
        if (curr_node.key < x):
            if (curr_node.right == None):
                curr_node.right = Node(x)
            else:
                self._insert(x, curr_node.right)
        else:
            if (curr_node.left == None):
                curr_node.left = Node(x)
            else:
                self._insert(x, curr_node.left)
        return True

def insert(self, x):
    if self.root == None:
        self.root = Node(x)
    else:
        self._insert(x, self.root)

BinarySearchTree._insert = _insert
BinarySearchTree.insert = insert
```

```python
bst = BinarySearchTree()
for i in range(10):
    bst.insert(i)

bst.inorder_traversal()
```

## 3) Heap

* CompleteBinaryTree 기반. 
* max/min 값을 찾기 위함
 * max heap: parent node.key > sibling node.key
 * min heap: parent node.key < sibling node.key

```python
from collections import deque

class Heap:
    def __init__(self):
        self.heap = deque()
        self.tail = -1
        
class MaxHeap(Heap):
    def __init__(self):
        super().__init__()
        
    def heapify(self):
        parents, sibling = int(self.tail/2) - 1, self.tail
        
        while parents >= 0:
            if self.heap[sibling] > self.heap[parents]:
                tmp = self.heap[parents]
                self.heap[parents] = self.heap[sibling]
                self.heap[sibling] = tmp
            sibling = parents
            parents = int(parents/2) - 1
            

class MinHeap(Heap):
    def __init__(self):
        super().__init__()
        
maxHeap = MaxHeap()
```

### 3-1) insertion

```python
def insert(self, x):
    self.heap.append(x)
    self.tail += 1
    self.heapify()
    
MaxHeap.insert = insert
```
