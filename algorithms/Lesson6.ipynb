## MergeSort
* time complexity = O(nlogn)
* algorithm:



```python
import numpy as np

LARGE_NUM = 1e+20
def Merge(A, p, q, r):
    n1 = q - p +1
    n2 = r - q
    L, R = [0] * (n1 + 1), [0]* (n2 + 1)
    for i in range(n1):
        L[i] = A[p + i]
    for j in range(n2):
        R[j] = A[q + j +1]

    L[n1], R[n2] = LARGE_NUM, LARGE_NUM
    
    i, j = 0, 0
    for k in range(p, r + 1):
        if L[i] < R[j]:
            A[k] = L[i]
            i = i + 1
        else:
            A[k] = R[j]
            j = j + 1
    return
    
def _MergeSort(A, p, r):
    if p < r:
        q = int((p + r) / 2)
        _MergeSort(A, p, q)
        _MergeSort(A, q+1, r)
        Merge(A, p, q, r)
    return

def MergeSort(A):
    _MergeSort(A, 0, len(A)-1)
    return


A = np.random.randint(low=0, high=1000, size=10)
print("Before Sorted: ", A)
MergeSort(A)
print("After Sorted: ", A)
```
