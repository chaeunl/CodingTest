### Solution 1: Stack

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        s = list(str(x))
        if "-" in s:
            return False
        else:
            stack = []
            s = list(str(x))
            half = len(s) / 2
        if len(s) % 2 != 0:
            s.pop(int(len(s)/2))
        while len(s) > half:
                stack.append(s.pop(0))   
        while len(stack) > 0:
            if stack.pop(-1) != s.pop(0):
                return False
            
        return True
```

```python
solution = Solution().isPalindrome
print(solution(x=121))
print(solution(x=-121))
print(solution(x=10))
```

### Solution 2: String

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        return str(x) == str(x)[::-1]
```

```python
solution = Solution().isPalindrome
print(solution(x=121))
print(solution(x=-121))
print(solution(x=10))
```

### Solution 3: Math w/o using string

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        orig = x
        back_x = 0
        while x > 0:
            back_x = (back_x * 10) + (x % 10)
            x = x // 10
        return orig == back_x
```

```python
solution = Solution().isPalindrome
print(solution(x=121))
print(solution(x=-121))
print(solution(x=10))
```
