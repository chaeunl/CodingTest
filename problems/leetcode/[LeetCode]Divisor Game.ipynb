### Solution 1: Mathematical proof

```python
class Solution:
    def divisorGame(self, n: int) -> bool:
        return (n % 2) == 0 
```

```python
solution = Solution().divisorGame

print(solution(n=2))
print(solution(n=3))
print(solution(n=1003))
```

### Solution 2: Dynamic programming

```python
class Solution(object):
    def divisorGame(self, n: int) -> bool:
        dp = [False for i in range(n + 1)] 
        dp[0] = True
        for i in range(1, n + 1):
             for j in range(1, i):
                if i % j == 0 and not dp[i - j]:
                    dp[i] = True
                    break
        return dp[n]
```

```python
solution = Solution().divisorGame

print(solution(n=2))
print(solution(n=3))
print(solution(n=1003))
```
