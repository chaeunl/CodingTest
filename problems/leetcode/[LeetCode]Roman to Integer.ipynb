```python
class Solution:
    def romanToInt(self, s: str) -> int:
        roman = {"M":1000, "D":500, "C":100, "L":50, "X":10, "V":5, "I":1}
        num = 0
        prev, curr = 0, 0
        for i in range(len(s)):
            curr = roman[s[i]]
            if prev < curr:
                num += curr - 2*prev
            else:
                num += curr
            prev = curr
        return num
```
