### Solution 1: hash map

```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        
        hash_map = {}
        
        for idx_a, a in enumerate(nums):
            b = target - a
            if a in hash_map:
                return [hash_map[a], idx_a]
            else:
                hash_map[b] = idx_a
        return []
```

```python
solution = Solution().twoSum
print(solution(nums=[2,7,11,15], target=9))
print(solution(nums=[3,2,4], target=6))
print(solution(nums=[3,3], target=6))
```

### Solution 2: double loop

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if target == nums[i] + nums[j]:
                    return [i,j]
```

```python
solution = Solution().twoSum
print(solution(nums=[2,7,11,15], target=9))
print(solution(nums=[3,2,4], target=6))
print(solution(nums=[3,3], target=6))
```
